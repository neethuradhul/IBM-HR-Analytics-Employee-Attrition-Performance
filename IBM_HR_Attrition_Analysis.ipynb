##  Exploratory Insights

### 1. Key Attrition Trends
- **Overall Attrition Rate:** Around *16%–20%* of employees have left the organization, indicating a moderate turnover level that warrants attention.
- **Departmental Impact:** The **Sales** and **Human Resources** departments experience higher attrition compared to **Research & Development (R&D)**. This suggests potential issues related to job pressure, incentives, or career growth in these teams.
- **Job Role Sensitivity:** **Laboratory Technicians, Sales Executives, and Human Resources roles** show the highest attrition levels — likely due to workload imbalance, limited progression, or external job opportunities.

### 2. Demographic Insights
- **Age Group:** Attrition is higher among employees aged **25–35 years**, representing early-career professionals who often explore better prospects.
- **Marital Status:** **Single employees** show a notably higher attrition rate, possibly due to mobility and flexibility to switch jobs.
- **Gender:** **Male attrition** is slightly higher than female attrition, potentially linked to departmental distributions or role expectations.

### 3. Job-Related Insights
- **Years at Company:** Employees with **<3 years of experience** at the company are more likely to leave, suggesting onboarding or early engagement issues.
- **Years in Current Role:** Attrition peaks among those in the same role for **2–5 years**, hinting at stagnation or lack of promotion opportunities.
- **OverTime:** A strong positive correlation exists between **OverTime and Attrition** — employees working frequent overtime hours are more likely to leave, likely due to burnout or work-life imbalance.
- **Job Satisfaction:** Lower satisfaction levels strongly correlate with attrition. Retention strategies should focus on improving role clarity, recognition, and work environment.

### 4. Compensation & Performance
- **Monthly Income:** Lower-income employees show **significantly higher attrition**, suggesting dissatisfaction with compensation or market competitiveness.
- **Stock Options & Benefits:** Employees with **no stock options** or lower benefit levels have higher attrition — financial incentives play a critical retention role.
- **Performance Rating:** No strong direct relationship observed between performance rating and attrition, implying that high performers may also be leaving — a concerning sign for talent retention.

### 5. Work Environment & Balance
- **Work-Life Balance:** Employees rating this factor as “Poor” or “Fair” have a much higher likelihood of leaving the organization.
- **Environment Satisfaction:** Teams with low environmental satisfaction (e.g., management support, resources, or interpersonal relations) exhibit higher turnover.

### 6. Key Drivers of Attrition
| Rank | Factor | Influence on Attrition |
|------|---------|------------------------|
| 1 | OverTime | Strong positive correlation |
| 2 | Monthly Income | Lower income → higher attrition |
| 3 | Years at Company | New employees leave more |
| 4 | Job Satisfaction | Lower satisfaction → higher attrition |
| 5 | Work-Life Balance | Poor balance → higher attrition |

### 7. Recommendations
- **Implement Early Retention Programs:** Focus on employees in their first 2 years with mentoring and career growth paths.
- **Optimize Workload Distribution:** Reduce overtime frequency through better resource planning.
- **Review Compensation Structures:** Ensure salaries are competitive for high-turnover roles, especially in Sales and HR.
- **Enhance Career Progression:** Introduce internal mobility and skill development programs.
- **Promote Work-Life Balance:** Encourage flexible hours or hybrid work models where possible.
- **Monitor High-Risk Groups:** Continuously track attrition patterns among young, single, and low-income employees.

---

### Conclusion
Employee attrition in this dataset is primarily driven by **workload pressure, compensation dissatisfaction, limited career growth, and early disengagement**. Addressing these through **targeted retention, rewards, and engagement initiatives** can significantly reduce turnover and enhance organizational stability.

## Preprocessing Steps Done:

3. Replaced numerical values in 'EnvironmentSatisfaction' column with corresponding categorical value (1 'Low' 2 'Medium' 3 'High' 4 'Very High')
4. Replaced numerical values in 'JobSatisfaction' column with corresponding categorical value (1 'Low' 2 'Medium' 3 'High' 4 'Very High')

6. Replaced numerical values in 'JobInvolvement' column with corresponding categorical value (1 'Low' 2 'Medium' 3 'High' 4 'Very High')

```python
# 1.  Checked null/missing values 
df.isnull().sum()
```

```python
# 2. Replaced numerical values in 'Education' column with corresponding categorical value (1 'Below College' 2 'College' 3 'Bachelor' 4 'Master' 5 'Doctor')
df['Education'] = df['Education'].map({
    1: 'Below College',
    2: 'College',
    3: 'Bachelor',
    4: 'Master',
    5: 'Doctor'
})
```

```python
# 3. Replaced numerical values in 'EnvironmentSatisfaction' column with corresponding categorical value (1 'Low' 2 'Medium' 3 'High' 4 'Very High')
df['EnvironmentSatisfaction'] = df['EnvironmentSatisfaction'].map({
    1: 'Low',
    2: 'Medium',
    3: 'High',
    4: 'Very High'
   })
```

```python
# 4. Replaced numerical values in 'JobSatisfaction' column with corresponding categorical value (1 'Low' 2 'Medium' 3 'High' 4 'Very High')
# df['JobSatisfaction'].value_counts()
df['JobSatisfaction'] = df['JobSatisfaction'].map({
    1: 'Low',
    2: 'Medium',
    3: 'High',
    4: 'Very High'
   })
```

```python
# 5. Attrition column values are converted to numeric values for better analysis
df['Attrition'] = df['Attrition'].map({'Yes': 1, 'No': 0})
```

```python
# 6. Replaced numerical values in 'JobInvolvement' column with corresponding categorical value (1 'Low' 2 'Medium' 3 'High' 4 'Very High')
df['JobInvolvement'] = df['JobInvolvement'].map({
    1: 'Low',
    2: 'Medium',
    3: 'High',
    4: 'Very High'
   })
```

```python
# 7. Replaced numerical values in 'PerformanceRating' column with corresponding categorical value (1 'Low' 2 'Good' 3 'Excellent' 4 'Outstanding')
df['PerformanceRating'] = df['PerformanceRating'].map({
    1: 'Low',
    2: 'Good',
    3: 'Excellent',
    4: 'Outstanding'
   })
```

```python
# 8. Replaced numerical values in 'RelationshipSatisfaction' column with corresponding categorical value (1 'Low' 2 'Medium' 3 'High' 4 'Very High')
df['RelationshipSatisfaction'] = df['RelationshipSatisfaction'].map({
    1: 'Low',
    2: 'Medium',
    3: 'High',
    4: 'Very High'
   })
```

```python
# 9. Replaced numerical values in 'WorkLifeBalance' column with corresponding categorical value (1 'Bad' 2 'Good' 3 'Better' 4 'Best')
df['WorkLifeBalance'] = df['WorkLifeBalance'].map({
    1: 'Poor',
    2: 'Fair',
    3: 'Good',
    4: 'Excellent'
   })
```

```python
# 10. Dropped 'EmployeeCount', 'StandardHours', 'Over18' columns as these columns values are constant and don’t provide useful insights for analysis 
df = df.drop(columns=['EmployeeCount', 'StandardHours', 'Over18','EmployeeNumber'])
```

## Exploratory Data Analysis

```python
# Purpose  : To show the overall attrition trend within the organization
# Insights : The pie chart shows that only a small proportion of employees (around 15–20%) have left the organization,
#             while the majority (about 80–85%) continue to stay.            
# Finding  : The overall attrition rate is moderate and retention is generally good, but a relatively low number of employees are leaving 
#             and it is still significant enough to impact productivity and recruitment costs.
#             Further analysis is required to identify which factors are driving attrition.

attrition = df['Attrition'].value_counts().reset_index()
attrition.columns = ['Attrition', 'Count']
plt.pie(attrition['Count'], labels=['No', 'Yes'], autopct='%1.1f%%', startangle=90, wedgeprops = {'edgecolor':'white'},
    colors =plt.cm.Set2.colors)
plt.title('Employee Attrition Distribution',fontsize=14)
plt.show()
```

```python
# Count plot showing overall attrition trend based on Gender 
# Purpose: To examine whether attrition rates differ between male and female employees.
# Insight: The proportions of attrition among male and female employees are nearly equal,
#           suggesting that gender has no major impact on employee turnover.
# Finding: Attrition is not gender-biased; both male and female employees have similar likelihoods of leaving.
#          Retention efforts can be gender-neutral and focused on work-related factors instead.

plt.figure(figsize=(8,6))
sns.countplot(x='Gender',hue=df['Attrition'].map({1: 'Yes', 0: 'No'}),palette = ['Red','Green'], data=df)
plt.title('Attrition by Gender', fontsize = 14)
plt.xlabel("Gender")
plt.ylabel("Count")
plt.legend(title='Attrition',bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()
```

```python
# Histogram showing destribution of employee age and attrition
# Purpose: To observe the overall age destribution of employees.
# Insight: Most employees fall between 25–40 years, showing a young to mid-career workforce.
# Finding: Retention efforts may need to target younger employees who are early in their careers.

plt.figure(figsize=(8,6))
sns.histplot(x='Age', kde=True, hue=df['Attrition'].map({1: 'Yes', 0: 'No'}), palette=['red','green'],bins=10, data=df)
plt.title("Distribution of Employee Age", fontsize = 14)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
# Count plot showing Department Wise Attrition Distribution
# Purpose: To identify departments with the highest attrition.
# Insight: Sales and HR departments show higher attrition than R&D.
# Finding: Sales employees may face higher pressure or fewer rewards — targeted retention policies are needed.

plt.figure(figsize=(8,6))
sns.countplot(x='Department',hue=df['Attrition'].map({1: 'Yes', 0: 'No'}),palette=['yellow','darkblue'], data=df)
plt.title('Department Wise Attrition Distribution',fontsize = 14)
plt.xlabel("Department")
plt.ylabel("Attrition Count")
plt.show()
```

```python
# Count plot showing Attrition by OverTime
# Purpose: To understand the impact of overtime on attrition.
# Insight: Employees who work OverTime show significantly higher attrition.
# Finding: Work-life imbalance is a major cause of turnover. Monitor workload and ensure fair time-off policies.

plt.figure(figsize=(8,6))
sns.countplot(x='OverTime',hue=df['Attrition'].map({1: 'Yes', 0: 'No'}),palette = 'flare', data=df)
plt.title('Attrition by OverTime', fontsize = 14)
plt.xlabel("OverTime")
plt.ylabel("Count")
plt.legend(title='Attrition',bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()
```

```python
# Purpose: To examine how WorkLifeBalance interacts with OverTime and its effect on employee attrition.
# Insight : Employees with low WorkLifeBalance who also work OverTime have the highest average attrition.
# Finding : OverTime is a key driver of low WorkLifeBalance and attrition.
#           HR can reduce turnover by limiting unnecessary overtime,Implementing flexible work hours,Promoting workload management.

lifebal_Ot = df.groupby(['WorkLifeBalance','OverTime'], as_index=False)['Attrition'].mean()

plt.figure(figsize=(8,6))
sns.barplot(x='WorkLifeBalance',y='Attrition',hue='OverTime',data=lifebal_Ot,palette=['deeppink','royalblue'])
plt.title('Average Attrition by WorkLife Balance and OverTime', fontsize=14)
plt.xlabel('WorkLifeBalance')
plt.ylabel('Average Employees Left')
plt.tight_layout()
plt.show()
```

```python
# Box plot showing the salary spread across employees
# Purpose: To view distribution and outliers in monthly income
# Insight: Income is right-skewed, meaning most employees earn below the mean, while a few earn significantly higher.
# Finding: There may be income inequality between job levels — a potential source of dissatisfaction.
#          The distribution is strongly right-skewed, there is a few high earners even after filtering,
#           could be legitimate high-income employees (Executives/Managers) that naturally fall beyond typical ranges.

plt.figure(figsize=(12,6))
Q1 = df['MonthlyIncome'].quantile(0.25)
Q3 = df['MonthlyIncome'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers_removed = df[(df['MonthlyIncome'] >= lower_bound) & (df['MonthlyIncome'] <= upper_bound)]
print(f"Removed {len(df) - len(outliers_removed)} outliers from MonthlyIncome.")

plt.subplot(1, 2, 1)
sns.boxplot(x='MonthlyIncome', color='lightcoral', data=df)
plt.title('Monthly Income (With Outliers)')
plt.xlabel('Monthly Income')

plt.subplot(1, 2, 2)
sns.boxplot(x='MonthlyIncome', color='mediumseagreen', data=outliers_removed)
plt.title('Monthly Income (Outliers Removed)')
plt.xlabel('Monthly Income')

plt.tight_layout()
plt.show()
```

```python
# Bar plotshowing Attrition by Job Role
# Purpose: To examine which job roles are most affected by attrition.
# Insight: Sales Representatives and Lab Technicians experience the highest attrition.
# Finding: Frontline and technical support roles are more vulnerable to turnover due to workload or limited advancement.

plt.figure(figsize=(10,6))
sns.barplot(x='JobRole', y='Attrition', estimator='mean',hue = 'JobRole',palette = 'mako',errorbar=None, data=df)
plt.title("Average Attrition by JobRole", fontsize=14)
plt.ylabel("Average Attrition")
plt.xlabel("JobRole")
labels = [textwrap.fill(label, 12) for label in df["JobRole"].unique()]
plt.xticks(ticks=range(len(labels)), labels=labels)
plt.tight_layout()
plt.show()
```

```python
# Count plot to visualize Attrition by Marital Status
# Purpose: To check if personal life factors influence attrition.
# Insight: Single employees show higher attrition.
# Finding: Employees without family commitments may have greater flexibility to switch jobs. Engagement activities could improve their retention.

plt.figure(figsize=(8,6))
sns.countplot(x='MaritalStatus',hue=df['Attrition'].map({1: 'Yes', 0: 'No'}),palette = ['darkslategray','aquamarine'], data=df)
plt.title('Attrition by MaritalStatus', fontsize = 14)
plt.xlabel("MaritalStatus")
plt.ylabel("Count")
plt.legend(title='Attrition',bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()
```

```python
# Attrition by Age & Monthly Income
# Purpose: To analyse how age and salary interact in influencing attrition.
# Insight: Younger employees with lower income show higher attrition.
# Finding: Compensation and early-career growth are critical retention drivers.

plt.figure(figsize=(8,6))
df_sample = df.sample(n=100)
sns.scatterplot(x='Age', y='MonthlyIncome', hue=df['Attrition'].map({1: 'Yes', 0: 'No'}), data=df_sample, palette='colorblind')
plt.title("Age vs Monthly Income")
plt.xlabel("Age")
plt.ylabel("Monthly Income")
plt.legend(title='Attrition',bbox_to_anchor=(1.05, 1), loc="upper left")
plt.show()
```

```python
# Count plot showing overall attrition trend based on Travel Frequency
# Purpose : To compare the attrition rates based on Travel Frequency and identify if based on Travel Frequency influences employee turnover.
# Insight: The count plot shows that employees who travel frequently have more proportion of attrition.
# Finding: Travel Frequency may to be a factor influencing employee attrition.

plt.figure(figsize=(8,6))
sns.countplot(x='BusinessTravel',hue=df['Attrition'].map({1: 'Yes', 0: 'No'}),palette = 'mako', data=df)
plt.title('Attrition by Business Travel Frequency', fontsize = 14)
plt.xlabel("Business Travel Frequency")
plt.ylabel("Count")
plt.legend(title='Attrition',bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()
```

```python
#  Purpose: To analyze whether Distance From Home influences employee attrition
# Insight : The boxplot indicates that employees who left generally lived farther from the company than those who stayed.
# Finding : However, if the medians are similar, DistanceFromHome may not be a significant factor in attrition for this dataset.

plt.figure(figsize=(8,5))
sns.boxplot(x='Attrition', y='DistanceFromHome', data=df,hue=df['Attrition'].map({1: 'Yes', 0: 'No'}) , palette=['indigo','palevioletred'])
plt.title('Attrition vs Distance From Home', fontsize=14)
plt.xlabel('Attrition')
plt.ylabel('Distance From Home (km)')
plt.legend(title='Attrition',bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()
```

```python
# Purpose : To measures financial incentives beyond base salary influence attrition
# Insight : Employees with StockOptionLevel = 0 (no options) often have higher attrition, especially at lower job levels.
#           Employees at higher levels (StockOptionLevel 2–3) are usually less likely to leave, as stock options increase financial commitment.
#           StockOptionLevel 3, employees may value career progression, job satisfaction, or work-life balance more than the financial incentive.
# Finding : Financial incentives like stock options, combined with career level, help retain employees.


stock_data = (df.groupby(['JobLevel', 'StockOptionLevel'])['Attrition'].mean().reset_index())
pivot_data = stock_data.pivot(index='JobLevel', columns='StockOptionLevel', values='Attrition')

plt.figure(figsize=(8,6))
sns.heatmap(pivot_data,annot=True,fmt=".2f",cmap='coolwarm',linewidths=0.5,cbar_kws={'label': 'Attrition Rate'})
plt.title('Attrition Rate by Job Level and Stock Option Level', fontsize=14)
plt.xlabel('Stock Option Level')
plt.ylabel('Job Level')
plt.tight_layout()
plt.show()
```

```python
# Histogram with KDE plot showing Distribution of Years at Company by Attrition Status
# Purpose : To analyse does employee turnover occur more frequently among new hires, mid-tenure staff, or long-tenure employees?
# Insight : Employees with fewer years at the company (typically < 3 years) have higher attrition rates.
#           The KDE curves flatten for higher tenure, showing very few long-tenured employees leave.
# Finding : Early departures indicate onboarding or role-fit issues.Strengthen onboarding, mentoring, and early engagement

plt.figure(figsize=(10,6))
sns.histplot(x='YearsAtCompany', hue=df['Attrition'].map({1: 'Yes', 0: 'No'}), kde=True, bins=20,data=df, palette='mako', alpha=0.6)
plt.title('Distribution of Years at Company by Attrition Status', fontsize=14)
plt.xlabel('Years at Company')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# bar plot shows average salary hike for each Department and Attrition 
# Purpose : To see whether salary hikes influence employee attrition 
# Insight : No major salary hike gap between employees who stayed and those who left in any department.
# Finding : This suggests that salary hikes alone do not significantly impact attrition in this dataset.Non-financial factors likely drive attrition

avg_hike = df.groupby(['Department', 'Attrition'], as_index=False)['PercentSalaryHike'].mean()

plt.figure(figsize=(10,6))
sns.barplot(x='Department',y='PercentSalaryHike',hue=avg_hike['Attrition'].map({1: 'Yes', 0: 'No'}),data=avg_hike,palette=['Navy', 'Yellow'])
plt.title('Average Percentage Salary Hike by Department and Attrition', fontsize=14)
plt.xlabel('Department')
plt.ylabel('Average Percent Salary Hike')
plt.legend(title='Attrition', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Purpose : To see whether salary hikes influence employee attrition 
# Insight : The median PercentSalaryHike is often similar between employees who left and those who stayed.
# Finding : This suggests that salary hikes alone do not significantly impact attrition .Non-financial factors likely drive attrition

plt.figure(figsize=(8,6))
sns.boxplot(x='Attrition',y='PercentSalaryHike',hue=df['Attrition'].map({1: 'Yes', 0: 'No'}),data=df,palette=['steelblue','crimson'])

plt.title('Percentage Salary Hike by Attrition', fontsize=14)
plt.xlabel('Attrition')
plt.ylabel('Percentage Salary Hike')
plt.legend(title='Attrition',bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()
```

```python
# Heat map shows Attrition by JobLevel,TotalWorkingYears and MonthlyIncome
# Purpose: To understand how experience, job level and Monthly income relate to attrition.
# Insight: Income, experience, and job level are positively related — they all grow together.
#          Attrition shows negative correlation with all three, that is Lower-level, less experienced,and lower-paid employees are more likely to leave.
#          The strong link between JobLevel and MonthlyIncome suggests that promotions may be a natural motivator for retention.
# Finding: Implementing clear promotion paths, competitive pay scales, and early engagement programs can help retain this segment.

corr_matrix = df[['Attrition', 'JobLevel', 'TotalWorkingYears','MonthlyIncome']].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True,  fmt=".2f", linewidths=0.5)
plt.title('Correlation: Attrition, Job Level, Total Working Years & Monthly Income', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# lineplot showing of years with current manager and Attrition
# Purpose : To analyze how the number of years an employee has worked under their current manager correlates with attrition rates.
# Insight : Employees with less than 2 years under the same manager show higher attrition.As the number of years with the same manager increases,
#            attrition tends to decrease, suggesting stability and stronger employee-manager relationships over time.
# Finding : Organizations can reduce attrition by ensuring better onboarding under new managers,
#           consistent feedback, and leadership training to build long-term employee engagement and reduce adjustment issues, lack of trust, or communication gaps.

counts = df.groupby(['YearsWithCurrManager','Attrition']).size().reset_index(name='EmployeeCount')

plt.figure(figsize=(10,6))
sns.lineplot(x='YearsWithCurrManager', y='EmployeeCount',hue=counts['Attrition'].map({1: 'Yes', 0: 'No'}),data=counts,marker='o',palette='Set1')
plt.title('Years With Current Manager at Company by Attrition')
plt.xlabel('Years With Current Manager')
plt.ylabel('Number of Employees')
plt.xticks(range(df['YearsWithCurrManager'].min(), df['YearsWithCurrManager'].max()+1))
plt.legend(title='Attrition', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Bar plot showing overall attrition trend based on Department and Travel Frequency
# Purpose : To compare the attrition rates based on Department and Travel Frequency and identify if it influences employee turnover.
# Insight: The bar plot shows that employees who travel frequently have more proportion of attrition in each department
# Finding: Travel Frequency appear to be a  major factor influencing employee attrition.

import textwrap
Attr_Dept = df.groupby(["Department","BusinessTravel"], as_index=False)["Attrition"].mean()

plt.figure(figsize=(10,6))
sns.barplot(x = 'Department',y = 'Attrition',hue = 'BusinessTravel',palette = 'Paired', data = Attr_Dept)
plt.title("Average Attrition by Department and Travel Frequency", fontsize=14)
plt.ylabel("Average Attrition")
plt.xlabel("Department")
labels = [textwrap.fill(label, 12) for label in df["Department"].unique()]
plt.xticks(ticks=range(len(labels)), labels=labels)
plt.legend(title='Travel Frequency',bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()
```

```python
# Purpose : To compare how the time since an employee’s last promotion and the time spent in their current role differ between those who left and those who stayed 
# Insight : Years Since Last Promotion - Employees who left tend to have higher median years since their last promotion.
#           Years in Current Role - Employees with long tenure in the same role show higher attrition rates.
# Finding: Lack of promotion for several years may lead to demotivation and resignation.Promotion is a strong driver of attrition.
#          Staying too long in the same position may reduce engagement and satisfaction.

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
sns.boxplot(x='Attrition', y='YearsSinceLastPromotion',hue='Attrition', data=df, palette=['steelblue','crimson'])
plt.title('Years Since Last Promotion by Attrition')
plt.xlabel('Attrition')
plt.ylabel('Years Since Last Promotion')

plt.subplot(1,2,2)
sns.boxplot(x='Attrition', y='YearsInCurrentRole', data=df, hue='Attrition',palette=['steelblue','crimson'])
plt.title('Years in Current Role by Attrition')
plt.xlabel('Attrition')
plt.ylabel('Years in Current Role')

plt.tight_layout()
plt.show()

```

```python
# scattre plot showing Job Satisfaction vs Years At Company by Attrition
# Purpose: To examine whether employee satisfaction increases with tenure and how this affects attrition patterns.
# Insight : Indicates that long-tenured employees are generally more satisfied and less likely to leave.
# Finding : Early-career employees with low satisfaction are high-risk for turnover.
sample_data = df.sample(n=100, random_state=42)

plt.figure(figsize=(10,6))
sns.scatterplot(x='YearsAtCompany',y='JobSatisfaction', hue=df['Attrition'].map({1: 'Yes', 0: 'No'}),data=sample_data,palette=['steelblue', 'crimson'])
plt.title("Job Satisfaction vs Years At Company by Attrition", fontsize=14)
plt.xlabel('Years at Company')
plt.ylabel('Job Satisfaction (1 = Low, 4 = High)')
plt.legend(title='Attrition', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

```

```python
# Violin plot showing Attrition by EnvironmentSatisfaction & JobSatisfaction
# Purpose : To check job and environmentsatisfaction levels impact attrition.
# Insight : The distribution of Attrition = Yes is heavier at Low and Medium Environment Satisfaction.
#           Employees with Very High Environment Satisfaction have very few leaving (Attrition = Yes).
# Finding : Focus on employees with Low or Medium Environment Satisfaction — improve workspace, culture, and facilities.
#           Target employees with Low Job Satisfaction — provide recognition, growth opportunities, and job clarity.

plt.figure(figsize=(12,5))

plt.subplot(1, 2, 1)
sns.violinplot(x='EnvironmentSatisfaction',y='Attrition',hue = 'EnvironmentSatisfaction',data=df,palette='rocket')
plt.title('Attrition by Environment Satisfaction', fontsize=13)
plt.xlabel('Environment Satisfaction')
plt.ylabel('Attrition ')


plt.subplot(1, 2, 2)
sns.violinplot(x='JobSatisfaction', y='Attrition',hue = 'JobSatisfaction',data=df,palette='mako')
plt.title('Attrition by Job Satisfaction', fontsize=13)
plt.xlabel('Job Satisfaction')
plt.ylabel('Attrition')
plt.tight_layout()
plt.show()

```

```python
# Correlation Heatmap - All Numeric Variables 
# Purpose: To visualize correlations among all numeric variables and identify which features are most strongly associated with employee attrition.
# Insight:  Attrition correlates negatively with MonthlyIncome, JobLevel, and YearsAtCompany.
#           It correlates positively with OverTime,DistanceFromHome

# Finding: Employees with higher salaries are less likely to leave.
#          Senior-level employees tend to have more stability and lower attrition.
#          Longer tenure corresponds to lower turnover, showing loyalty and experience retention.
#          Employees working overtime are significantly more likely to quit, possibly due to burnout or workload stress.Suggesting work-life imbalance is a major retention issue.
#          Employees living farther away may experience commuting stress, slightly raising attrition risk.
#          Factors like PerformanceRating and TrainingTimesLastYear show weak or no relationship with attrition,meaning these variables might not directly impact turnover.

numeric_values = df.select_dtypes(include=['int64', 'float64'])
corr = numeric_values.corr()

plt.figure(figsize=(12,8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap of Numeric Features', fontsize=14)
plt.show()
```

```python
# Pairplot visualizes pairwise relationships among key numeric features — Age, MonthlyIncome, and YearsAtCompany
# Purpose: To highlights how these relationships differ between employees who left versus those who stayed.
# Insight: Employees with short tenure and lower income tend to cluster under Attrition = Yes.
#          Higher-tenured employees with higher income usually have Attrition = No, indicating retention increases with experience and compensation.
#          Age alone may have less visible impact, but combined with tenure and income, patterns emerge.
# Finding: Tenure (YearsAtCompany) and MonthlyIncome are strong predictors of attrition.
#          HR can focus on early-career employees with low pay for retention programs.

sample_data = df.sample(n=100, random_state=42)
sns.pairplot(sample_data[['Age', 'MonthlyIncome', 'YearsAtCompany', 'Attrition']], hue='Attrition')
plt.suptitle('Pairwise Relationships Age, MonthlyIncome, YearsAtCompany,Attrition', y=1.02)
plt.show()
```

##  Key Insights and Findings

- **Overall Attrition Rate:** Around *16%–20%* of employees have left the organization, indicating a moderate turnover level that warrants attention.
- **Departmental Impact:** The **Sales** and **Human Resources** departments experience higher attrition compared to **Research & Development (R&D)**. This suggests potential issues related to job pressure, incentives, or career growth in these teams.
- **Job Role Sensitivity:** **Laboratory Technicians, Sales Executives, and Human Resources roles** show the highest attrition levels — likely due to workload imbalance, limited progression, or external job opportunities.


- **Age Group:** Attrition is higher among employees aged **25–35 years**, representing early-career professionals who often explore better prospects.
- **Marital Status:** **Single employees** show a notably higher attrition rate, possibly due to mobility and flexibility to switch jobs.
- **Gender:** **Male attrition** is slightly higher than female attrition, potentially linked to departmental distributions or role expectations.


- **Years at Company:** Employees with **<3 years of experience** at the company are more likely to leave, suggesting onboarding or early engagement issues.
- **Years in Current Role:** Attrition peaks among those in the same role for **2–5 years**, hinting at stagnation or lack of promotion opportunities.
- **OverTime:** A strong positive correlation exists between **OverTime and Attrition** — employees working frequent overtime hours are more likely to leave, likely due to burnout or work-life imbalance.
- **Job Satisfaction:** Lower satisfaction levels strongly correlate with attrition. Retention strategies should focus on improving role clarity, recognition, and work environment.


- **Monthly Income:** Lower-income employees show **significantly higher attrition**, suggesting dissatisfaction with compensation or market competitiveness.
- **Stock Options & Benefits:** Employees with **no stock options** or lower benefit levels have higher attrition — financial incentives play a critical retention role.
- **Performance Rating:** No strong direct relationship observed between performance rating and attrition, implying that high performers may also be leaving — a concerning sign for talent retention.


- **Work-Life Balance:** Employees rating this factor as “Poor” or “Fair” have a much higher likelihood of leaving the organization.
- **Environment Satisfaction:** Teams with low environmental satisfaction (e.g., management support, resources, or interpersonal relations) exhibit higher turnover.
